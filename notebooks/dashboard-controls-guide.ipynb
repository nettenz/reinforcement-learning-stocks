## Control Panel Breakdown

### **1. Movement threshold (default: 0.00)**
- **Purpose**: Defines the minimum price movement percentage to classify as "actionable"
- **Current**: 0.00 (captures all movements, even tiny ones)
- **Effect**: Higher values (e.g., 0.002 = 0.2%) filter out small noise movements
- **Recommendation**: Try 0.002-0.005 for meaningful signal evaluation

### **2. Prediction horizon (steps) (default: 1)**
- **Purpose**: How many steps ahead to evaluate if the signal was "correct"
- **Current**: 1 step = next day prediction
- **Effect**: Longer horizons (2-5) test if signals predict multi-day trends
- **Use case**: 1 for day trading, 3-5 for swing trading analysis

### **3. Deterministic policy (argmax action) ✓**
- **Purpose**: Forces model to always pick the highest-probability action
- **Current**: ON (deterministic)
- **Effect**: OFF would add randomness/exploration during signal generation
- **Recommendation**: Keep ON for consistent signal evaluation

### **4. Chart window (latest rows) (default: 1000)**
- **Purpose**: How many recent data points to display in charts
- **Current**: 1000 rows = ~3 years of daily data
- **Effect**: Lower values zoom into recent periods, higher shows more history
- **Performance**: Reduce if charts load slowly

### **5. Show Buy/Sell text labels ○**
- **Purpose**: Toggles text annotations on buy/sell markers in price charts
- **Current**: OFF (cleaner visual)
- **Effect**: ON adds "BUY"/"SELL" text labels to chart markers
- **Use case**: Turn ON for detailed signal inspection

### **6. Signal label density (default: 12)**
- **Purpose**: Controls how many signal annotations appear on charts
- **Current**: 12 (moderate density)
- **Effect**: Higher = more labels, Lower = fewer labels for cleaner view
- **Tip**: Reduce to 5-8 for busy charts, increase to 20+ for detailed analysis

### **7. Show horizon-return panel ✓**
- **Purpose**: Displays panel showing forward returns for each signal
- **Current**: ON
- **Effect**: Shows if buy/sell signals actually led to profitable moves
- **Value**: Essential for validating signal quality

### **8. Highlight incorrect actionable signals ✓**
- **Purpose**: Visually emphasizes signals that were wrong about market direction
- **Current**: ON (helpful for debugging)
- **Effect**: Makes failed signals stand out in red/different color
- **Use case**: Critical for identifying model weaknesses

## Recommended Settings by Use Case

### **General Analysis**
```
Movement threshold: 0.002 (filter noise)
Horizon: 1-3 (depending on strategy timeframe)
Chart window: 500-1000 (balance detail vs performance)
Signal density: 8-15 (readable but informative)
All toggles: ON (comprehensive analysis)
```

### **Performance Debugging**
```
Movement threshold: 0.005 (focus on significant moves)
Horizon: 1 (immediate accuracy)
Chart window: 500 (recent performance)
Signal density: 20 (show all signals)
Highlight incorrect: ON (find failure patterns)
```

### **Clean Overview**
```
Movement threshold: 0.002
Horizon: 3 (swing trading perspective)
Chart window: 1000 (broad context)
Signal density: 8 (minimal clutter)
Show labels: OFF (clean charts)
```
